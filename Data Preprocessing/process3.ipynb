```python
import requests
import json
import requests
import numpy as np
from tmdbv3api import Movie, TMDb
tmdb=TMDb()
tmdb.api_key='fbe6f1f741a0c62d9394cf6f5f76b65f'
tmdb_movie=Movie()
def get_genre(x):
    genres=[]
    result=tmdb_movie.search(x)
    print(result)
    movie_id=result[0].id
    print(movie_id)
    url = "https://api.themoviedb.org/3/movie/{}?api_key=fbe6f1f741a0c62d9394cf6f5f76b65f&language=en-US".format(movie_id)
    response= requests.get(url)
    # timeout=(10, 30))
    data_json = response.json()
    print(data_json)
    if data_json['genres']:
        genre_str = " "
        for i in range(0, len(data_json['genres'])):
            print(f"{i} {data_json['genres'][i]}")
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        print("None")
        np.NaN
```

```python
df['Genre'] = df['Title'].map(lambda x:get_genre(str(x)))
```
