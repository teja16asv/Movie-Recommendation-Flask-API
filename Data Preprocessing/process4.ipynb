```python
import pandas as pd
import numpy as np
import requests
import bs4 as bs
import urllib.request
```

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2020"
```

```python
source=urllib.request.urlopen(link).read()
soup=bs.BeautifulSoup(source,'lxml')
```

```python
tables=soup.find_all('table', class_='wikitable sortable')
print(len(tables))
```

```python
type(tables[0])
```

```python
df1 = pd.read_html(str(tables[0]))[0]
df2 = pd.read_html(str(tables[1]))[0]
df3 = pd.read_html(str(tables[2]))[0]
df4 = pd.read_html(str(tables[3]).replace("'1\"\'",'"1"'))[0]
```

```python
df=pd.concat([df1,df2,df3,df4], ignore_index=True)
```

```python
df
```

```python
df_2020=df[['Title','Cast and crew']]
print(df_2020)
```

```python
from tmdbv3api import TMDb, Movie
import json
import requests
import numpy as np
tmdb=TMDb()
tmdb.api_key = 'f260a69950e1011c06058a3a956d76f6'
tmdb_movie = Movie() 
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
      return np.NaN
    else:
      try:
        print('result: \n',result)
        print('result[0]: \n',result[0])
        movie_id = result[0].id
        print('movie_id: ',movie_id)
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key=f260a69950e1011c06058a3a956d76f6'.format(movie_id))
        data_json = response.json()
        print('data_json: \n',data_json)
        print("--------------------")
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            return np.NaN
      except Exception as e:
        return np.NaN
```

```python
df_2020['genres']=df_2020['Title'].map(lambda x: get_genre(str(x)))
```

```python
df_2020
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2020.loc[:,'director_name'] = df_2020['Cast and crew'].map(lambda x: get_director(str(x)))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2020.loc[:,'actor_1_name'] = df_2020['Cast and crew'].map(lambda x: get_actor1(str(x)))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2020.loc[:,'actor_2_name'] = df_2020['Cast and crew'].map(lambda x: get_actor2(str(x)))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2020.loc[:,'actor_3_name'] = df_2020['Cast and crew'].map(lambda x: get_actor3(str(x)))
```

```python
df_2020=df_2020.rename(columns={'Title':'movie_title'})
```
